## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import numpy as np
import pandas as pd

!wget https://raw.githubusercontent.com/gazizovaa/run-the-starter-notebooks/refs/heads/main/data/raw/content_refresh_anonymized.csv
```

```python
df = pd.read_csv("content_refresh_anonymized.csv")
print(f"Total number of rows: {len(df)}")
print()

# declining pages have different average impressions than others
trend_pages = df.groupby('trend_direction')["impressions_90d"].mean()
print(trend_pages)
print()

# CTR differ meaningfully by position tier
ctr_meaning = df.groupby("position_tier")["ctr"].mean()
print(ctr_meaning)
print()

# how many pages are both visible (enough impressions) and stale?
stale_visible = df[(df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)]
print("Stale and visible pages:", len(stale_visible), f"({len(stale_visible)/len(df):.1%} of dataset)")
```
